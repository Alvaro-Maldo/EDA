```python
MovieUrl = 'http://www.imdb.com/chart/toptv'

response = requests.get(MovieUrl)
soup = BeautifulSoup(response.text, 'lxml')
link_=soup.find_all(class_="titleColumn")
links_=[]
title=[]

for id_ in link_:
    
    final_url=id_.find("a").get("href") 
    url="http://imdb.com" + final_url
    links_.append(url)

links_

for i in links_:

    headers = {"Accept-Language": "en-US,en;q=0.5"}
    response1 = requests.get(i, headers=headers)
    soup1 = BeautifulSoup(response1.text, 'html.parser')

    title_=soup1.find("div", class_="sc-80d4314-1 fbQftq").h1.get_text()



    print(title_)
    title.append(title_)


```

```python
IMDBS["title"]=title
IMDBS
```

```python
ratings=soup.find_all(class_="posterColumn")
rates_=[]

for indexa in ratings:

    rate=round(float(indexa.find("span", attrs={"name":"ir"})["data-value"]),2)
    rates_.append(rate)


IMDBS["Rates"]=rates_
IMDBS
```

```python
ratings=soup.find_all(class_="ratingColumn imdbRating")
nvotes=[]
for index in ratings:

    votes=index.find("strong").get("title").split("on")[1].split(" ")[1].replace(",","")
    nvotes.append(int(votes))


nvotes

IMDBS["NVotes"]=nvotes

IMDBS
```

```python
IMDBS.to_csv("IMDBS", sep='\t')
```
